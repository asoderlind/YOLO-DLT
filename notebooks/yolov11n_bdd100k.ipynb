```python
from ultralytics import YOLO

# Create a new YOLO model from scratch
# model = YOLO("yolo11n.yaml")

# Load a pretrained YOLO model (recommended for training)
model = YOLO("yolo11n.pt")
```

```python
results = model.train(data="../ultralytics/cfg/datasets/bdd10k.yaml", epochs=1, device="mps")
```

```python

# Perform object detection on an image using the model
results = model(f"{dataset_path}/images/val/b1f6c103-8b75ea3e.jpg")
for result in results:
    image_with_boxes = result.plot()  # Returns an image with bounding boxes (NumPy array)

    # Convert BGR (OpenCV) to RGB (Matplotlib format)
    image_with_boxes = cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB)

    # Display the image inline
    plt.figure(figsize=(10, 6))  # Set figure size
    plt.imshow(image_with_boxes)
    plt.axis("off")  # Hide axes
    plt.show()
```

```python

def draw_yolo_bboxes(image_path, label_path):
    # Load the image
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    with open(label_path, 'rt') as f:
        data_lines = f.readlines()
    
    bboxes = []
    w, h = 1280, 720

    for item in range(len(data_lines)):
        cl=int(data_lines[item].split(" ")[0])
        a=float(data_lines[item].split(" ")[1])
        b=float(data_lines[item].split(" ")[2])
        c=float(data_lines[item].split(" ")[3])
        d=float(data_lines[item].split(" ")[4][:-1])
        
        x1 = int((a-c/2)*w)
        y1 = int((b-d/2)*h)
        x2 = int((a+c/2)*w)
        y2 = int((b+d/2)*h)
        
        bboxes.append([x1,y1,x2,y2,cl])

    for c in bboxes:
        cv2.rectangle(img, (c[0],c[1]), (c[2], c[3]), (0, 255, 0), 5)
        cv2.putText(img, str(c[4]), (int((c[0]+c[2])/2), int((c[1]+c[3])/2)), 0, 1, (0, 0, 255),3)
        
    plt.figure(figsize=(10, 6))  # Set figure size
    plt.imshow(img)
    plt.axis("off")  # Hide axes
    plt.show()


# Example usage
img_path = f"{dataset_path}/images/val/b1f6c103-8b75ea3e.jpg"
label_path = f"{dataset_path}/labels/val/b1f6c103-8b75ea3e.txt"

draw_yolo_bboxes(img_path, label_path)
```
