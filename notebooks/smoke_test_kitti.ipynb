```python
import random
import glob
import os
import PIL

from utils import draw_yolo_bboxes

dataset_path = "../../yolo-testing/datasets/kitti-yolo"

class2index = {
    'Car': 0,
    'Van': 1,
    'Truck': 2,
    'Pedestrian': 3,
    'Person_sitting': 4,
    'Cyclist': 5,
    'Tram': 6,
    'Misc': 7,
    'DontCare': 8
}

id2cls = {v: k for k, v in class2index.items()}

# Example usage
img_path = random.sample(glob.glob(f"{dataset_path}/images/train/*.png"), 1)[0]
image_name = os.path.basename(img_path)[:-4]
label_path = f"{dataset_path}/labels/train/{image_name}.txt"

print(img_path, label_path)

# get img_w and img_h from the image
img = PIL.Image.open(img_path)
img_w, img_h = img.size
print(img_w, img_h)

draw_yolo_bboxes(img_path, label_path, img_w, img_h, id2cls)
```
